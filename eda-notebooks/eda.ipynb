# Import

### Needed Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_palette('muted')
```

### Datasets

```python
ratings = pd.read_csv('data/driver_ratings.csv')
feedbacks = pd.read_csv('data/feedback.csv')
telemetry = pd.read_csv('data/telemetry.csv')
```

```python
for name, df in {'ratings':ratings, 'feedbacks':feedbacks, 'telemetry':telemetry}.items():
    print(f"Nulls found in \'{name}\' dataframe: {df.isna().sum().sum()}")
```

##### Data Type Checks

```python
ratings.info()
```

```python
feedbacks.info()
```

```python
feedbacks['timestamp'] = pd.to_datetime(feedbacks['timestamp'])
```

```python
telemetry.info()
```

From the dataframe information, we can cast certain features to a more appropriate type. `timestamp` can be converted to a datetime, and categorical features, such as `event_code`, `traffic_condition`, and `weather_condition`, can be casted to categorical data type (only after checking the cardinality).

`trip_duration_sec` can also converted to minutet measure instead, which is easier tot interpret. 

```python
telemetry['timestamp'] = pd.to_datetime(telemetry['timestamp'])


telemetry['trip_duration_min'] = telemetry['trip_duration_sec'] / 60
telemetry.drop(columns=['trip_duration_sec'], inplace=True)
```

```python
categoricals = ['event_code','traffic_condition','weather_condition']
```

Checking feature cardinality for the categorical columns. Since they are mostly limited to at most five categories, we can convert them to a categorical data type to save some space. 

```python
for col in categoricals:
    print(f"Unique vals. for col. \'{col}\':")
    print(telemetry[col].unique())
    print()
```

```python
# normalizing traffic and weather condition to title case for readability
# event_code was left unmodified (except for data type) for now
telemetry['traffic_condition'] = telemetry['traffic_condition'].str.title()
telemetry['weather_condition'] = telemetry['weather_condition'].str.replace('CLR','Clear')
telemetry['weather_condition'] = telemetry['weather_condition'].str.title()

for col in categoricals:
    telemetry[col] = telemetry[col].astype('category')

telemetry['traffic_condition'] = telemetry['traffic_condition'].cat.reorder_categories(['Light','Moderate','Heavy'], ordered=True)
```

##### Distribution Checks

As per usual, data distributions should be checked in order to determine the underlying structure of the data. This allows us to spot data anomalies, choose the appropriate transformations (if any), and avoid making incorrect assumptions/conclusions. 

Starting with the ratings table first. 

```python
ratings.head()
```

The rating column is the only continuous numerical feature that can be examined in this table. The values can be put plotted using a histogram to reveal how most customers rate the delivery services and a boxplot to show the range of ratings across all deliveries.

The ratings appear to be left-skewed with most having values around the higher 4.0s. Notice in the histogram that there is another peak at the 5.0 mark, which shows that there are over 2500 customers who have had very satisfactory delivery experiences. This is also due to the bounds of the rating system, where the scale only goes up to 5.0 inclusive. Moreover, there are about less than around 400 customers who have given mid to low ratings with the lowest rating recorded being around 2.0. Overall, these plots show that customers are generally satisfied with the delivery services and only a few have had unpleasant experiences.

Note that the four columns representing acceleration in x, y, and z dimensions are signed. We can replace them with the absolute magnitude instead, since the column name itself already denotes the direction of the force. This also makes correlations easier to interpret later on. 

```python
cols = ['peak_fwd_x_g',
        'peak_rear_x_g',
        'peak_left_y_g',
        'peak_right_y_g',
        'peak_up_z_g',
        'peak_down_z_g'
]

for col in cols:
    telemetry[col] = telemetry[col].abs()
```

```python
fig, ax = plt.subplots(2,2, figsize=(13, 6))

cols = ['speed_kmh', 'max_steering_rate_deg_per_s']

sns.boxplot(
    data=telemetry,
    x=cols[0],
    ax=ax[0,0],
)
sns.boxplot(
    data=telemetry,
    x=cols[1],
    ax=ax[0,1]
)

sns.histplot(
    data=telemetry,
    x=cols[0],
    ax=ax[1,0]
)
sns.histplot(
    data=telemetry,
    x=cols[1],
    ax=ax[1,1]
)


plt.tight_layout()
plt.show()
```

Next, we examine average speed and maximum steering rate per second. Most delivery drivers operate the vehicles with an average speed of around 20-60 km/h, implying adherence to speed limits and traffic rules. There also appears to be a small amount of observations where delivery vehicles travel way slower than 20km/h. This may be due to situations such as periods of heavy traffic jams, prompting drivers to slow down and frequently stop. The right tail also shows that only a few delivery vehicles achieved an average speed in the higher 80's. A likely reason for this is that such vehicles travelled using expressways and drove at the speed limit (which is around 80 km/h for trucks and buses). The boxplot also shows that there are observations at the far right of the distribution. These indicate that there are some drivers that frequently go over the speed limit in certain areas. 

As for maximum steering rate per second, the data appears to look normally distributed around 150 deg/s. Not much can be concluded from looking at these values alone to determine risky driving behavior. Trips with high steering rates can have no record of swerving, because such steering input can be made when performing parking maneuvers. The same can be said for relatively lower steering rates. At high speeds, even an input of 100 deg/s can be dangerous since the vehicle's speed amplifies lateral forces, increasing the chance of swerving/sliding. 

```python
fig, ax = plt.subplots(3, 2, figsize=(12, 7))

box_cols = [
    'peak_fwd_x_g',
    'peak_rear_x_g',
    'peak_left_y_g',
    'peak_right_y_g',
    'peak_up_z_g',
    'peak_down_z_g',
]

row_idx = 0
col_idx = 0

for col in box_cols:

    if col_idx > 1:
        row_idx += 1
        col_idx = 0

    sns.histplot(
        data=telemetry,
        x=col,
        ax=ax[row_idx, col_idx]
    )

    col_idx += 1

plt.tight_layout()
plt.subplots_adjust(hspace=0.5, wspace=0.2)
plt.show()
```

The distributions for longitudinal and lateral acceleration exhibit the same right skew. The observations are pulled slightly to the left, indicating that most drivers are smooth on the braking and steering. Of course, on the right tail, there are a few trips with g-forces approaching the highehr 0.7-0.8g's, which is quite high for vehicles such as delivery trucks/vans. These values may indicate aggressive handling (hard braking and steering) or accident avoidance (braking hard or steering away from obstacles such as jaywalking pedestrians, other vehicles, etc.).

There is also a distinct peak in the data distribution for vertical g-force. A high count of deliveries with these close-to-zero values likely mean that the routes followed by most delivery vehicles have little to no hills or inclined roads. It could also mean that drivers are taking extra caution when driving on inclines, as most vertical acceleration values cluster around 0 to 0.15g, with maximums reaching up to 0.3g.

#### Feature Engineering

Rating labels can be added to the ratings table, where each rating is assignend to a specific bin. This allows raw rating values to be categorized. 

```python
ratings['rating_category'] = pd.cut(
    ratings['rating'],
    bins=[1.0, 2.5, 3.5, 4.2, 5.0],
    labels=['Poor','Fair','Good','Excellent']
)
```

We also create features for peak lateral and longitudinal accelerations, since both represent only one type of magnitude. This also reduces simplifies our correlations (and reduces clutter in the heatmap later on). 

```python
telemetry['peak_long_g'] = telemetry[['peak_fwd_x_g', 'peak_rear_x_g']].max(axis=1)
telemetry['peak_lat_g'] = telemetry[['peak_right_y_g','peak_left_y_g']].max(axis=1)
```

Finally, we create columns for the weekday and the time of day deliveries were made. 

```python
telemetry['delivery_dayofweek'] = telemetry['timestamp'].dt.day_name()
telemetry['time_of_day'] = pd.cut(
    telemetry['timestamp'].dt.hour,
    bins=[7, 11, 15, 20],
    labels=['Morning', 'Noon/Early Afternoon','Late Afternoon/Evening']
)
```

##### Correlations

We can then examine the correlations between numeric features. Since the ratings table contain a numeric value too (rating), we merge it together with the telemetry table. 

```python
merged = telemetry.merge(ratings, on=['trip_id','driver_id'])

telemetry_ratings_merge = merged[["rating", 
        "late_delivery_flag", 
        "overspeed_count", 
        "swerving_count",
        "hard_brake_count",
        "speed_kmh",
        "handling_risk_score",
        "distance_tr_km",
        "max_steering_rate_deg_per_s",
        "peak_long_g",
        "peak_lat_g",
]]
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

sns.heatmap(
    telemetry_ratings_merge.corr('spearman',numeric_only=True), 
    annot=True,
    ax=ax
)

plt.show()
```

We use Spearman's rank correlation to calculate the correlation coefficients, since most of the features have a skewed shape. 

At first glance, we notice a group of moderate to strong pairwise correlations at the lower right portion of the heatmap. These correlations are mostly telemetry data interactions which do not provide us with much information on driver behavior. 

Also notice `overspeed_count`, `swerving_count`, and `hard_brake_count` have moderate to strong correlations with `handling_risk_score`. We can infer which feature influences this score the most later on. 

Apart from this, the correlation between `rating` and `handling_risk_score` shows a moderate correlation of about -0.42. Since `handling_risk_score` indicates how likely packages are to get damaged, we can assume that the more aggressive a driver is with their handling, the lower the customers' satisfaction will be due to package damage. 

There is also a strong correlation between `rating` and `late_delivery_flag`, which simply means that a late delivery usually gets lower ratings. 

```python
fig, ax = plt.subplots(1, 4, figsize=(17, 5))

data=telemetry_ratings_merge.sample(20000, random_state=42)

sns.stripplot(
    data=data,
    x='swerving_count',
    y='rating',
    jitter=True,
    alpha=0.25,
    ax=ax[0]
)

sns.stripplot(
    data=data,
    x='hard_brake_count',
    y='rating',
    jitter=True,
    alpha=0.25,
    ax=ax[1]
)


sns.stripplot(
    data=data,
    x='overspeed_count',
    y='rating',
    jitter=True,
    alpha=0.25,
    ax=ax[2]
)

sns.stripplot(
    data=data,
    x='late_delivery_flag',
    y='rating',
    jitter=True,
    alpha=0.25,
    ax=ax[3]
)


plt.tight_layout()
plt.show()
```

We use a jittered strip plot on the supposedly 'constituent' features of `handling_risk_score` and the `late_delivery_flag`, and compare each to `rating`. Notice the drop in ratings as swerving, hard brarking, and overspeed instances increase. The most significant looking drop occurs for `swerving_count` and the least for `overspeed_count`. This shows that vehicle swerves are the primary cause of compromised package integrity, resulting in lower ratings. 

#### General Aggregates

```python
fig, ax = plt.subplots(2, 2, figsize=(19, 9))

sns.countplot(
    data=telemetry.loc[telemetry['event_code'] != 'NONE',:],
    y='event_code',
    order=['SWERV','HARD_BRK','OVRSPD','MULTI'],
    ax=ax[0, 0]
)

sns.countplot(
    data=merged.loc[merged['late_delivery_flag'] == 1,:],
    x='traffic_condition',
    hue='weather_condition',
    ax=ax[0, 1]
)

sns.countplot(
    data=merged.assign(late_delivery_flag=merged['late_delivery_flag'].map({1:'Late',0:'On time'})),
    order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],
    x='delivery_dayofweek',
    hue='late_delivery_flag',
    ax=ax[1, 0]
)

sns.countplot(
    data=merged.assign(late_delivery_flag=merged['late_delivery_flag'].map({1:'Late',0:'On time'})),
    x='time_of_day',
    hue='late_delivery_flag',
    ax=ax[1, 1]
)

ax[0, 1].set_ylabel('late_deliveries')
plt.tight_layout()
plt.show()
```

The above plots describe miscellaneous details and general insights from the data.
